```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,20)
y1 = x**2
y2 = x**1.5
plt.plot(x,y1,"bo-",linewidth=2, markersize=6, label="First")
plt.plot(x,y2,"ro-",linewidth=2, markersize=6, label="Second")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axis([-0.5,10.5,-5,105])
plt.legend(loc="upper left")
plt.savefig("myplot.jpg")
```
