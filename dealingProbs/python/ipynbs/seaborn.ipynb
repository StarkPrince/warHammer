```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.columns
```

```python
sns.displot(auto['price'],kde=False,rug=True)
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind='hex')
```

```python
sns.jointplot(x=auto['engine_size'],y=auto['horsepower'],kind='kde')
```

```python
sns.pairplot(auto[['engine_size','normalized_losses','compression_ratio','peak_rpm','price']])
```

```python
sns.swarmplot(x=auto['fuel_type'],y=auto['horsepower'])
```

```python
sns.stripplot(x=auto['fuel_type'],y=auto['horsepower'],jitter=True)
```

```python
sns.boxplot(x=auto['number_of_doors'],y=auto['horsepower'],hue=auto['make'])
```
